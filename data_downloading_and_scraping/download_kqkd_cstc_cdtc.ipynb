# Download data

```python
def download_data(stock_symbol: str):
    bank_stock = Vnstock().stock(symbol=stock_symbol, source="VCI")
    bank_kqkd = bank_stock.finance.income_statement(period="quarter", lang="vi", dropna=True)
    bank_cstc = bank_stock.finance.ratio(period="quarter", lang="vi", dropna=True)
    bank_cdtc= bank_stock.finance.balance_sheet(period="quarter", lang="vi", dropna=True)
    return bank_kqkd, bank_cstc, bank_cdtc
```

# Select columns and time

```python
def filter_time(bank_dataframe: pd.DataFrame):
    bank_dataframe = bank_dataframe[
        (bank_dataframe["Năm"] >= 2021) &
        (bank_dataframe["Năm"] != 2021) |
        ((bank_dataframe["Năm"] == 2021) & (bank_dataframe["Kỳ"] == 4))
    ]
    return bank_dataframe

def select_columns_and_time(
    bank_kqkd: pd.DataFrame,
    bank_cstc: pd.DataFrame,
    bank_cdtc: pd.DataFrame
):
    bank_kqkd = bank_kqkd[[
        "Năm",
        "Kỳ",
        "Doanh thu (đồng)",
        "Tăng trưởng doanh thu (%)",
        "Lợi nhuận sau thuế của Cổ đông công ty mẹ (đồng)",
        "Tăng trưởng lợi nhuận (%)",
        "LN trước thuế",
        "Lợi nhuận thuần",
        "Cố tức đã nhận"
    ]]
    bank_kqkd = filter_time(bank_kqkd)

    bank_cstc = bank_cstc[["Meta", "Chỉ tiêu khả năng sinh lợi", "Chỉ tiêu định giá"]]
    bank_cstc = bank_cstc.droplevel(0, axis=1)
    bank_cstc = filter_time(bank_cstc)

    bank_cdtc = bank_cdtc[[
        "Năm",
        "Kỳ",
        "TỔNG CỘNG TÀI SẢN (đồng)",
        "TỔNG CỘNG NGUỒN VỐN (đồng)",
        "NỢ PHẢI TRẢ (đồng)",
        "VỐN CHỦ SỞ HỮU (đồng)",
        "LỢI ÍCH CỦA CỔ ĐÔNG THIỂU SỐ"
    ]]
    bank_cdtc = filter_time(bank_cdtc)

    return bank_kqkd, bank_cstc, bank_cdtc
```
