# Import modules

```python
from vnstock import Vnstock
```

# Download data

```python
def download_trade(stock_symbol: str):
    bank_stock = Vnstock().stock(symbol=stock_symbol, source="VCI")
    bank_trade = bank_stock.quote.history(start="2022-01-01", end="2025-03-31", interval="1D")
    return bank_trade
```

```python
vcb_trade = download_trade("VCB")
vcb_trade
```

```python
bid_trade = download_trade("BID")
bid_trade
```

```python
tcb_trade = download_trade("TCB")
tcb_trade
```
