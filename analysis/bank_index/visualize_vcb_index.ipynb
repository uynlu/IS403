## Kết quả kinh doanh

```python
vcb_kqkd["Tháng"] = vcb_kqkd["Kỳ"] * 3
vcb_kqkd
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=vcb_kqkd["Năm"].astype(str) + "-" + vcb_kqkd["Tháng"].astype(str),
    y=vcb_kqkd["Doanh thu (đồng)"],
    mode="lines+markers",
    name="Doanh thu (đồng)"
))

fig.update_layout(
    title="Doanh thu (đồng) theo thời gian",
    xaxis_title="Năm - Kỳ",
    yaxis_title="Doanh thu (đồng)"
)

fig.show()
```

Theo biểu đồ, ta thấy:
  - Từ đầu năm 2022 đến quý 2 năm 2023, doanh thu có xu hướng tăng và đạt đỉnh ở quý 2 năm 2023.
  - Sau quý 2 năm 2023, doanh thu giảm dần, sau đó phục hồi dần từ quý 2 năm 2024.

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=vcb_kqkd["Năm"].astype(str) + "-" + vcb_kqkd["Tháng"].astype(str),
    y=vcb_kqkd["Tăng trưởng doanh thu (%)"],
    marker_color=["green" if val > 0 else "red" for val in vcb_kqkd["Tăng trưởng doanh thu (%)"]]
))

fig.update_layout(
    title="Tăng trưởng doanh thu (%) theo thời gian",
    xaxis_title="Năm - Kỳ",
    yaxis_title="Tăng trưởng doanh thu (%)"
)

fig.show()
```

Theo biểu đồ tăng trưởng doanh thu, ta thấy:
  - Từ đầu năm 2022 - quý 3 năm 2023, tăng trưởng doanh thu dương. Trong đó, quý 1 năm 2023 tăng trưởng doanh thu đạt đỉnh.
  - Trong năm 2022, tăng trưởng doanh thu có xu hướng tăng.
  - Từ năm 2023, tăng trưởng doanh thu có xu hướng giảm. Trong đó từ cuối năm 2023 tăng trưởng doanh thu bắt đầu âm và đạt đáy vào quý 2 năm 2024.

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=vcb_kqkd["Năm"].astype(str) + "-" + vcb_kqkd["Tháng"].astype(str),
    y=vcb_kqkd["Tăng trưởng lợi nhuận (%)"],
    marker_color=["green" if val > 0 else "red" for val in vcb_kqkd["Tăng trưởng lợi nhuận (%)"]]
))

fig.update_layout(
    title="Tăng trưởng lợi nhuận (%) theo thời gian",
    xaxis_title="Năm - Kỳ",
    yaxis_title="Tăng trưởng lợi nhuận (%)"
)

fig.show()
```

- Từ năm 2022 đến quý 3 năm 2023, tăng trưởng lợi nhuận dương, trong đó quý 4 năm 2022 tăng trưởng đạt đỉnh.
- Tăng trưởng lợi nhuận từ năm 2023 có xu hướng giảm, trong đó từ cuối năm 2023 đến đầu năm 2024 tăng trưởng âm.

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=vcb_kqkd["Năm"].astype(str) + "-" + vcb_kqkd["Tháng"].astype(str),
    y=vcb_kqkd["LN trước thuế"],
    mode="lines+markers",
    name="Lợi nhuận trước thuế (đồng)"
))
fig.add_trace(go.Scatter(
    x=vcb_kqkd["Năm"].astype(str) + "-" + vcb_kqkd["Tháng"].astype(str),
    y=vcb_kqkd["Lợi nhuận thuần"],
    mode="lines+markers",
    name="Lợi nhuận thuần (đồng)"
))

fig.update_layout(
    title="Lợi nhuận trước thuế và lợi nhuận thuần theo thời gian",
    xaxis_title="Năm - Kỳ",
    yaxis_title="Giá trị (đồng)"

)
fig.show()
```

## Chỉ số tài chính

```python
vcb_cstc["Tháng"] = vcb_cstc["Kỳ"] * 3
vcb_cstc
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
    y=vcb_cstc["ROE (%)"],
    mode="lines+markers",
    name="ROE (%)",
    yaxis="y1"
))
fig.add_trace(go.Scatter(
    x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
    y=vcb_cstc["ROA (%)"],
    mode="lines+markers",
    name="ROA (%)",
    yaxis="y2"))

fig.update_layout(
    title="ROE và ROA theo thời gian",
    xaxis_title="Năm - Kỳ",
    yaxis=dict(
        title="ROE (%)",
        tickfont=dict(color="#1f77b4")
    ),
    yaxis2=dict(
        title="ROA (%)",
        overlaying="y",
        side="right",
        tickfont=dict(color="#ff7f0e")
    )
)

fig.show()
```

Đối với ROE:
- Từ 2022: ROE có xu hướng tăng trưởng đáng kể trong suốt năm 2022, đạt đỉnh vào khoảng cuối năm 2022.
- Năm 2023 - 2024: Sau khi đạt đỉnh, ROE lại có xu hướng giảm mạnh.

Đối với ROA:
- Đầu năm 2022: ROA cũng cho thấy sự tăng trưởng nhẹ.
- Từ cuối năm 2022 - 2024: ROA có xu hướng chu kỳ, sau đó giảm mạnh vào cuối năm 2024.

Công ty  ban đầu dùng đòn bẩy tài chính hợp lý, nhưng về sau có dấu hiệu mất kiểm soát tài chính hoặc lợi nhuận sụt giảm trong khi nợ vẫn cao (do ROE tụt xuống dưới ROA).

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Bar(
        x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
        y=vcb_cstc["BVPS (VND)"],
        name="BVPS"
    ),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(
        x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
        y=vcb_cstc["EPS (VND)"],
        mode="lines+markers",
        name="EPS"
    ),
    secondary_y=True
)

fig.update_layout(
    title="BVPS và EPS theo thời gian",
    xaxis_title="Năm - Kỳ",
    yaxis_title="BVPS",
    yaxis2=dict(title="EPS", overlaying="y", side="right")
)

fig.show()
```

Về BVPS:
- BVPS có xu hướng tăng nhẹ theo chu kỳ.
- Mức tăng/giảm không quá lớn nhưng cho thấy giá trị tài sản ròng trên mỗi cổ phiếu có xu hướng đi lên.

Về EPS:
- EPS có sự biến động đáng kể hơn so với BVPS.
- Từ năm 2022, EPS có sự biến động mạnh. EPS đạt đỉnh vào khoảng cuối năm 2022.
- Từ năm 2023 trở đi, EPS có xu hướng giảm dần và đạt đáy vào cuối năm 2024.

```python
fig = make_subplots(
    rows=3,
    cols=1,
    shared_xaxes=True,
    vertical_spacing=0.03,
    subplot_titles=("P/E theo thời gian", "P/B theo thời gian", "P/S theo thời gian")
)

fig.add_trace(
    go.Scatter(
        x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
        y=vcb_cstc["P/E"], name="P/E"
    ),
    row=1,
    col=1
)
fig.add_trace(
    go.Scatter(
        x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
        y=vcb_cstc["P/B"], name="P/B"
    ),
    row=2,
    col=1
)
fig.add_trace(
    go.Scatter(
        x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
        y=vcb_cstc["P/S"], name="P/S"
    ),
    row=3,
    col=1
)

fig.update_layout(height=800, width=800, title_text="P/E, P/B, và P/S Ratios")
fig.update_xaxes(tickangle=-45, tickfont=dict(size=10), row=3, col=1)

fig.show()
```

Nhận xét về biểu đồ P/E:
- Biến động: Tỷ số P/E có sự biến động khá lớn trong giai đoạn này.
- Đỉnh và đáy: Có thể thấy nhiều đỉnh và đáy, cho thấy sự thay đổi trong kỳ vọng của thị trường về lợi nhuận của công ty.
- Xu hướng chung: Không có một xu hướng tăng hoặc giảm rõ rệt trong toàn bộ giai đoạn, mà là sự dao động lên xuống. Tuy nhiên, có vẻ như P/E có xu hướng tăng nhẹ vào cuối giai đoạn được thể hiện.

Nhận xét về biểu đồ P/B:
- Biến động: Tỷ số P/B cũng cho thấy sự biến động, nhưng có vẻ ít hơn so với P/E.
- Tăng trưởng và giảm: P/B có xu hướng tăng từ giữa năm 2022 và đạt đỉnh đầu năm 2023, sau đó giảm mạnh và đạt đáy vào cuối năm 2024.

Nhận xét về biểu đồ P/S:
- Xu hướng tăng rõ rệt: Tỷ số P/S cho thấy một xu hướng tăng khá rõ rệt từ giữa năm 2022 đến năm 2024.
- Ổn định vào năm 2024: Tỷ số P/S loanh quanh ở giá trị 7.5

Trong khi P/S có xu hướng tăng, P/E và P/B lại cho thấy sự biến động với các đỉnh và đáy khác nhau. Điều này có thể phản ánh sự thay đổi trong kỳ vọng của thị trường về các khía cạnh khác nhau của công ty (lợi nhuận, giá trị tài sản, doanh thu).

```python
fig = go.Figure(data=[go.Bar(
    x=vcb_cstc["Năm"].astype(str) + "-" + vcb_cstc["Tháng"].astype(str),
    y=vcb_cstc["P/Cash Flow"],
    marker_color=["red" if value < 0 else "green" for value in vcb_cstc["P/Cash Flow"]]
)])

fig.update_layout(
    title="P/Cash Flow theo thời gian",
    xaxis_title="Năm - Tháng",
    yaxis_title="P/Cash Flow",
    xaxis_tickangle=-45
)

fig.show()
```

Dựa trên biểu đồ, ta thấy:
- Đỉnh điểm bất thường: Vào tháng 6 năm 2024, tỷ số P/Cash Flow tăng vọt lên mức 123.35. Đây là một giá trị rất cao, cho thấy nhà đầu tư sẵn sàng trả một mức giá rất cao cho mỗi đơn vị dòng tiền của công ty vào thời điểm đó. Cần xem xét kỹ lưỡng các yếu tố nào đã dẫn đến sự tăng đột biến này.
- Xu hướng không rõ ràng: Không có một xu hướng tăng hoặc giảm rõ rệt nào của tỷ số P/Cash Flow trong suốt giai đoạn này, mà là sự biến động thất thường.
