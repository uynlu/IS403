## Bank index

```python
def set_time(dataframe: pd.DataFrame):
    def set_time_row(row):
        if row["Kỳ"] == 4:
            row["Năm"] = int(row["Năm"] + 1)
            row["Kỳ"] = int(1)
        else:
            row["Kỳ"] = int(row["Kỳ"] + 1)
        return row

    dataframe = dataframe.apply(set_time_row, axis=1)
    dataframe["Năm"] = dataframe["Năm"].astype(int)
    dataframe["Kỳ"] = dataframe["Kỳ"].astype(int)

    return dataframe
```

# Merge data

## Data

```python
tcb_dataset = pd.merge(tcb_trade, gdp, left_on="Quarter", right_on="Time", how="outer")
tcb_dataset.drop(columns=["Time_y"], inplace=True)
tcb_dataset.columns = ["Time"] + [column for column in tcb_dataset.columns[1:]]
tcb_dataset
```

```python
tcb_dataset = pd.merge(tcb_dataset, cpi, left_on="Month", right_on="Time", how="outer")
tcb_dataset.drop(columns=["Month", "Time_y"], inplace=True)
tcb_dataset.columns = ["Time"] + [column for column in tcb_dataset.columns[1:]]
tcb_dataset
```

```python
tcb_dataset = pd.merge(tcb_dataset, tcb_foreign_exchange_rate, left_on="Time", right_on="Date", how="left")
tcb_dataset.drop(columns=["Date"], inplace=True)
tcb_dataset
```

```python
tcb_dataset = pd.merge(tcb_dataset, tcb_kqkd, on="Quarter", how="left")
tcb_dataset.drop(columns=["Năm", "Kỳ"], inplace=True)
tcb_dataset
```

```python
tcb_dataset = pd.merge(tcb_dataset, tcb_cstc, on="Quarter", how="left")
tcb_dataset.drop(columns=["Năm", "Kỳ", "CP"], inplace=True)
tcb_dataset
```

```python
tcb_dataset = pd.merge(tcb_dataset, tcb_cdtc, on="Quarter", how="left")
tcb_dataset.drop(columns=["Năm", "Kỳ", "Quarter"], inplace=True)
tcb_dataset
```

## Future data

```python
tcb_future = tcb_trade[tcb_trade["Quarter"] == gdp.iloc[-1, :]["Time"]].iloc[-30:, :].copy()
tcb_future["Previous_GDP"] = gdp.iloc[-1, :]["Previous_GDP"]
tcb_future = tcb_future.reset_index().drop(columns=["index"])
tcb_future
```

```python
tcb_future = pd.merge(tcb_future, cpi, left_on="Month", right_on="Time", how="inner")
tcb_future.drop(columns=["Month", "Time_y"], inplace=True)
tcb_future.columns = ["Time"] + [column for column in tcb_future.columns[1:]]
tcb_future
```

```python
tcb_future = pd.merge(tcb_future, tcb_foreign_exchange_rate, left_on="Time", right_on="Date", how="inner")
tcb_future.drop(columns=["Date"], inplace=True)

last_row = tcb_future.iloc[-1]

new_row = {
    "Time": "2025-04-01",
    "Open": np.nan,
    "High": np.nan,
    "Low": np.nan,
    "Close": np.nan,
    "Volume": np.nan,
    "Quarter": last_row["Quarter"],
    "Previous_GDP": last_row["Previous_GDP"],
    "Previous_CPI": last_row["Previous_CPI"]
}

for columns in tcb_foreign_exchange_rate.columns:
    if columns.startswith("Previous_") and columns not in new_row:
        new_row[columns] = tcb_foreign_exchange_rate.iloc[-1][columns]

tcb_future = pd.concat([tcb_future, pd.DataFrame([new_row])], ignore_index=True)
tcb_future
```

```python
tcb_future = pd.merge(tcb_future, tcb_kqkd, on="Quarter", how="left")
tcb_future.drop(columns=["Năm", "Kỳ"], inplace=True)
tcb_future
```

```python
tcb_future = pd.merge(tcb_future, tcb_cstc, on="Quarter", how="left")
tcb_future.drop(columns=["Năm", "Kỳ", "CP"], inplace=True)
tcb_future
```

```python
tcb_future = pd.merge(tcb_future, tcb_cdtc, on="Quarter", how="left")
tcb_future.drop(columns=["Năm", "Kỳ", "Quarter"], inplace=True)
tcb_future
```
