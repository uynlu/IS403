## Bank index

```python
def set_time(dataframe: pd.DataFrame):
    def set_time_row(row):
        if row["Kỳ"] == 4:
            row["Năm"] = int(row["Năm"] + 1)
            row["Kỳ"] = int(1)
        else:
            row["Kỳ"] = int(row["Kỳ"] + 1)
        return row

    dataframe = dataframe.apply(set_time_row, axis=1)
    dataframe["Năm"] = dataframe["Năm"].astype(int)
    dataframe["Kỳ"] = dataframe["Kỳ"].astype(int)

    return dataframe
```

### Kết quả kinh doanh

## Trade

# Merge data

```python
bid_dataset = pd.merge(bid_trade, gdp, left_on="Quarter", right_on="Time", how="outer")
bid_dataset.drop(columns=["Time_y"], inplace=True)
bid_dataset.columns = ["Time"] + [column for column in bid_dataset.columns[1:]]
bid_dataset
```

```python
bid_dataset = pd.merge(bid_dataset, cpi, left_on="Month", right_on="Time", how="outer")
bid_dataset.drop(columns=["Month", "Time_y"], inplace=True)
bid_dataset.columns = ["Time"] + [column for column in bid_dataset.columns[1:]]
bid_dataset
```

```python
bid_dataset = pd.merge(bid_dataset, bid_foreign_exchange_rate, left_on="Time", right_on="Date", how="left")
bid_dataset.drop(columns=["Date"], inplace=True)
bid_dataset
```

```python
bid_dataset = pd.merge(bid_dataset, bid_kqkd, on="Quarter", how="left")
bid_dataset.drop(columns=["Năm", "Kỳ"], inplace=True)
bid_dataset
```

```python
bid_dataset = pd.merge(bid_dataset, bid_cstc, on="Quarter", how="left")
bid_dataset.drop(columns=["Năm", "Kỳ", "CP"], inplace=True)
bid_dataset
```

```python
bid_dataset = pd.merge(bid_dataset, bid_cdtc, on="Quarter", how="left")
bid_dataset.drop(columns=["Năm", "Kỳ", "Quarter"], inplace=True)
bid_dataset
```
