## Bank index

```python
def set_time(dataframe: pd.DataFrame):
    def set_time_row(row):
        if row["Kỳ"] == 4:
            row["Năm"] = int(row["Năm"] + 1)
            row["Kỳ"] = int(1)
        else:
            row["Kỳ"] = int(row["Kỳ"] + 1)
        return row

    dataframe = dataframe.apply(set_time_row, axis=1)
    dataframe["Năm"] = dataframe["Năm"].astype(int)
    dataframe["Kỳ"] = dataframe["Kỳ"].astype(int)

    return dataframe
```

# Merge data

## Data

```python
vcb_dataset = pd.merge(vcb_trade, gdp, left_on="Quarter", right_on="Time", how="outer")
vcb_dataset.drop(columns=["Time_y"], inplace=True)
vcb_dataset.columns = ["Time"] + [column for column in vcb_dataset.columns[1:]]
vcb_dataset
```

```python
vcb_dataset = pd.merge(vcb_dataset, cpi, left_on="Month", right_on="Time", how="outer")
vcb_dataset.drop(columns=["Month", "Time_y"], inplace=True)
vcb_dataset.columns = ["Time"] + [column for column in vcb_dataset.columns[1:]]
vcb_dataset
```

```python
vcb_dataset = pd.merge(vcb_dataset, vcb_foreign_exchange_rate, left_on="Time", right_on="Date", how="left")
vcb_dataset.drop(columns=["Date"], inplace=True)
vcb_dataset
```

```python
vcb_dataset = pd.merge(vcb_dataset, vcb_kqkd, on="Quarter", how="left")
vcb_dataset.drop(columns=["Năm", "Kỳ"], inplace=True)
vcb_dataset
```

```python
vcb_dataset = pd.merge(vcb_dataset, vcb_cstc, on="Quarter", how="left")
vcb_dataset.drop(columns=["Năm", "Kỳ", "CP"], inplace=True)
vcb_dataset
```

```python
vcb_dataset = pd.merge(vcb_dataset, vcb_cdtc, on="Quarter", how="left")
vcb_dataset.drop(columns=["Năm", "Kỳ", "Quarter"], inplace=True)
vcb_dataset
```

## Future data

```python
vcb_future = vcb_trade[vcb_trade["Quarter"] == gdp.iloc[-1, :]["Time"]].iloc[-30:, :].copy()
vcb_future["Previous_GDP"] = gdp.iloc[-1, :]["Previous_GDP"]
vcb_future = vcb_future.reset_index().drop(columns=["index"])
vcb_future
```

```python
vcb_future = pd.merge(vcb_future, cpi, left_on="Month", right_on="Time", how="inner")
vcb_future.drop(columns=["Month", "Time_y"], inplace=True)
vcb_future.columns = ["Time"] + [column for column in vcb_future.columns[1:]]
vcb_future
```

```python
vcb_future = pd.merge(vcb_future, vcb_foreign_exchange_rate, left_on="Time", right_on="Date", how="inner")
vcb_future.drop(columns=["Date"], inplace=True)

last_row = vcb_future.iloc[-1]

new_row = {
    "Time": "2025-04-01",
    "Open": np.nan,
    "High": np.nan,
    "Low": np.nan,
    "Close": np.nan,
    "Volume": np.nan,
    "Quarter": last_row["Quarter"],
    "Previous_GDP": last_row["Previous_GDP"],
    "Previous_CPI": last_row["Previous_CPI"]
}

for columns in vcb_foreign_exchange_rate.columns:
    if columns.startswith("Previous_") and columns not in new_row:
        new_row[columns] = vcb_foreign_exchange_rate.iloc[-1][columns]

vcb_future = pd.concat([vcb_future, pd.DataFrame([new_row])], ignore_index=True)
vcb_future
```

```python
vcb_future = pd.merge(vcb_future, vcb_kqkd, on="Quarter", how="left")
vcb_future.drop(columns=["Năm", "Kỳ"], inplace=True)
vcb_future
```

```python
vcb_future = pd.merge(vcb_future, vcb_cstc, on="Quarter", how="left")
vcb_future.drop(columns=["Năm", "Kỳ", "CP"], inplace=True)
vcb_future
```

```python
vcb_future = pd.merge(vcb_future, vcb_cdtc, on="Quarter", how="left")
vcb_future.drop(columns=["Năm", "Kỳ", "Quarter"], inplace=True)
vcb_future
```
