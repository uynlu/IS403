# Visualize

## Kết quả kinh doanh

```r
bid_kqkd$Tháng <- bid_kqkd$Kỳ * 3
head(bid_kqkd, 5)
```

```r
bid_kqkd$ThoiGian <- as.Date(paste0(bid_kqkd$Năm, "-", bid_kqkd$Tháng, "-01"))

ggplot(bid_kqkd, aes(x = ThoiGian, y = Doanh.thu..đồng.)) +
  geom_line(group = 1, color = "steelblue") +
  geom_point(color = "darkblue") +
  labs(
    title = "Doanh thu (đồng) theo thời gian",
    x = "Năm - Kỳ",
    y = "Doanh thu (đồng)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ, ta thấy:
- Từ đầu năm 2022 - 2025, doanh thu có xu hướng tăng và đạt đỉnh ở quý 2 năm 2023. Giai đoạn tăng mạnh nhất từ quý 2 2022 - quý 1 năm 2023.
- Sau quý 2 năm 2023, doanh thu có xu hướng giảm nhẹ. Nhưng từ đầu năm 2024, doanh thu bắt đầu phục hồi.

```r
bid_kqkd$color <- ifelse(bid_kqkd$Tăng.trưởng.doanh.thu.... > 0, "green", "red")

ggplot(bid_kqkd, aes(x = ThoiGian, y = Tăng.trưởng.doanh.thu.... , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Tăng trưởng doanh thu (%) theo thời gian",
    x = "Năm - Kỳ",
    y = "Tăng trưởng doanh thu (%)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ tăng trưởng doanh thu, ta thấy:
- Từ đầu năm 2022 - 2023, tăng trưởng doanh thu luôn dương. Trong đó, quý 2 năm 2023 tăng trưởng doanh thu đạt đỉnh.
- Trong năm 2022 - quý 2 2023, tăng trưởng doanh thu cao nhưng ổn định.
- Từ quý 2 năm 2023, tăng trưởng doanh thu có xu hướng giảm. Và bắt đầu tăng trưởng âm vào năm 2024, trong đó đạt đáy vào quý 2 2024. Sau đó tăng trưởng doanh thu có dấu hiệu phục hồi.

```r
bid_kqkd$color <- ifelse(bid_kqkd$Tăng.trưởng.lợi.nhuận.... > 0, "green", "red")

ggplot(bid_kqkd, aes(x = ThoiGian, y = Tăng.trưởng.lợi.nhuận.... , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Tăng trưởng lợi nhuận (%) theo thời gian",
    x = "Năm - Kỳ",
    y = "Tăng trưởng lợi nhuận (%)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ tăng trưởng lợi nhuận, ta thấy:
- Từ năm 2022 đến quý 3 năm 2023, tăng trưởng lợi nhuận dương, trong đó quý 4 năm 2022 tăng trưởng đạt đỉnh.
- Tăng trưởng lợi nhuận từ năm 2023 có xu hướng giảm, và tăng trưởng âm vào quý 3 2023.
- Tăng trưởng lợi nhuận từ cuối năm 2023 - 2024 không ổn định và không có xu hướng rõ ràng.

```r
data_long <- bid_kqkd %>%
  pivot_longer(cols = c(LN.trước.thuế, Lợi.nhuận.thuần),
              names_to = "Variable",
              values_to = "Value")

ggplot(data_long, aes(x = ThoiGian, y = Value, color = Variable)) +
  geom_line() +
  labs(title = "Lợi nhuận trước thuế và lợi nhuận thuần theo thời gian", x = "Năm - Kỳ", y = "Giá trị (đồng)") +
  theme_minimal()
```

Theo biểu đồ, ta thấy:
- Năm 2022 -2024, cả lợi nhuận trước thuế và sau thuế đều có xu hướng tăng, và có tính chu kì (tăng vào đầu năm và sụt giảm vào cuối năm).
- Năm 2023, lợi nhuận trước thuế và sau thuế ổn định quanh mức 4-5 nghìn tỉ đồng.
- Năm 2024, lợi nhuận trước thuế và sau thuế có biến động mạnh và đạt đỉnh vào cuối năm 2024.

```r
bid_kqkd$color <- ifelse(bid_kqkd$Cố.tức.đã.nhận > 0, "green", "red")

ggplot(bid_kqkd, aes(x = ThoiGian, y = Cố.tức.đã.nhận , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Cổ tức đã nhận theo thời gian",
    x = "Năm - Kỳ",
    y = "Cổ tức đã nhận"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

## Chỉ số tài chính

```r
bid_cstc$Tháng <- bid_cstc$Kỳ * 3
head(bid_cstc, 5)
```

```r
bid_cstc$ThoiGian <- as.Date(paste0(bid_cstc$Năm, "-", bid_cstc$Tháng, "-01"))

scale_factor <- max(bid_cstc$ROE...., na.rm = TRUE) / max(bid_cstc$ROA...., na.rm = TRUE)

ggplot(bid_cstc, aes(x = ThoiGian)) +
  geom_line(aes(y = ROE...., color = "ROE (%)"), size = 1) +
  geom_point(aes(y = ROE...., color = "ROE (%)")) +
  geom_line(aes(y = ROA.... * scale_factor, color = "ROA (%)"), size = 1) +
  geom_point(aes(y = ROA.... * scale_factor, color = "ROA (%)")) +
  scale_y_continuous(
    name = "ROE (%)",
    sec.axis = sec_axis(~ . / scale_factor, name = "ROA (%)")
  ) +
  scale_color_manual(
    name = "",
    values = c("ROE (%)" = "#1f77b4", "ROA (%)" = "#ff7f0e")
  ) +
  labs(
    title = "ROE và ROA theo thời gian",
    x = "Năm - Kỳ"
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    legend.position = "top"
  )
```

Theo biểu đồ, ta thấy đối với ROE:
- Từ năm 2022 - 2023, ROE tăng mạnh.
- Trong năm 2023: ROE có xu hướng giảm dần và đi ngang ở năm 2024, đặc biệt là từ giữa năm 2024 trở đi ROE duy trì ở mức khá cao, dao động quanh 0.19 - 0.20.

Đối với ROA:
- Năm 2022 - giữa năm 2023 ROA tăng mạnh, sau đó có sự điều chỉnh nhẹ ở quý 3 năm 2023 rồi quay lại mức cũ.
- Năm 2024: ROA duy trì ở mức tương đối ổn định quanh 0.01.

Có sự chênh lệch không lớn giữa ROA và ROE cho thấy ngân hàng đang sử dụng đòn bẩy tài chính ở mức an toàn.


```r
bid_cstc$color <- ifelse(bid_cstc$Biên.lợi.nhuận.ròng.... > 0, "green", "red")

ggplot(bid_cstc, aes(x = ThoiGian, y = Biên.lợi.nhuận.ròng.... , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Biên lợi nhuận ròng (%) theo thời gian",
    x = "Năm - Kỳ",
    y = "Biên lợi nhuận ròng (%)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

- Năm 2023 - 2024: Biên lợi nhuận ổn định quanh mức 0.4, tuy có sự suy giảm vào quý 3 của các năm. Nhưng biên lợi nhuận đã tăng và đạt mức cao nhất là vào cuối năm 2024 ở mức 0.5.

```r
bid_cstc$ThoiGian <- as.Date(paste0(bid_cstc$Năm, "-", bid_cstc$Tháng, "-01"))

data_long <- bid_cstc %>%
  pivot_longer(cols = c(Vốn.hóa..Tỷ.đồng.),
              names_to = "Variable",
              values_to = "Value")

ggplot(data_long, aes(x = ThoiGian, y = Value, color = Variable)) +
  geom_line() +
  labs(title = "Vốn hóa (tỷ đồng) theo thời gian", x = "Năm - Kỳ", y = "Vốn hóa (tỷ đồng)") +
  theme_minimal()
```

Theo biểu đồ, ta thấy:

- Từ sau quý 2 năm 2022: Vốn hoá có xu hướng tăng. Trong đó giai đoạn từ quý 2 2022 - quý 3 2023, vốn hoá tăng đều từ khoảng 89 nghìn tỷ lên 122 nghìn tỷ. Từ sau quý 3 2023 - quý 4 2024, vốn hoá tăng mạnh và có biến động mạnh.

```r
bid_cstc$color <- ifelse(bid_cstc$Số.CP.lưu.hành..Triệu.CP. > 0, "green", "red")

ggplot(bid_cstc, aes(x = ThoiGian, y = Số.CP.lưu.hành..Triệu.CP. , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Số cổ phiếu lưu hành (triệu cổ phiếu) theo thời gian",
    x = "Năm - Kỳ",
    y = "Số cổ phiếu lưu hành (triệu cổ phiếu)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Từ biểu đồ, ta thấy số cổ phiếu lưu hành có 2 lần điều chỉnh vào quý 3 năm 2023 và quý 3 năm 2024.

```r
bid_cstc$ThoiGian <- as.Date(paste0(bid_cstc$Năm, "-", bid_cstc$Tháng, "-01"))

scale_factor <- max(bid_cstc$BVPS..VND., na.rm = TRUE) / max(bid_cstc$EPS..VND., na.rm = TRUE)

ggplot(bid_cstc, aes(x = ThoiGian)) +
  geom_col(aes(y = BVPS..VND., fill = "BVPS"), width = 25) +
  geom_line(aes(y = EPS..VND. * scale_factor, color = "EPS"), size = 1.2) +
  geom_point(aes(y = EPS..VND. * scale_factor, color = "EPS")) +
  scale_y_continuous(
    name = "BVPS (VND)",
    sec.axis = sec_axis(~ . / scale_factor, name = "EPS (VND)")
  ) +
  scale_fill_manual(values = c("BVPS" = "skyblue"), name = "") +
  scale_color_manual(values = c("EPS" = "darkorange"), name = "") +
  labs(
    title = "BVPS và EPS theo thời gian",
    x = "Thời gian"
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    legend.position = "top"
  )
```

Theo biểu đồ, ta thấy:

Về BVPS:
- BVPS có xu hướng tăng trưởng khá ổn định và liên tục trong suốt giai đoạn khảo sát.
- Quý 1/2022 - Quý 4/2023: BVPS có xu hướng tăng trưởng khá ổn định trong giai đoạn này. Từ khoảng 28 nghìn đồng vào đầu năm 2022, BVPS tăng dần qua các quý và đạt mức cao nhất khoảng 37 nghìn đồng vào cuối năm 2023.
- Quý 1/2024 - Quý 4/2025: BVPS giảm xuống mức 19 nghìn đồng, sau đó có dấu hiệu đi ngang và duy trì ở mức tương đương, dao động nhẹ quanh mốc 20 nghìn đồng.


Về EPS:
- Năm 2022: EPS có sự biến động lớn trong giai đoạn này. Ban đầu ở mức khá cao, EPS giảm mạnh xuống mức thấp nhất vào cuối năm 2022.
- Năm 2023: EPS cho thấy sự phục hồi rõ rệt và có xu hướng ổn định hơn, dao động quanh mức 1250-1300.
- Từ đầu năm 2024, EPS bắt đầu có xu hướng giảm dần và giảm mạnh vào cuối năm 2024.

So sánh BVPS và EPS:
- BVPS cho thấy sự tăng trưởng ổn định trong phần lớn giai đoạn, cho thấy giá trị tài sản ròng trên mỗi cổ phiếu của công ty đang tăng lên.
- EPS lại có sự biến động mạnh hơn. Giai đoạn đầu có sự sụt giảm đáng kể, sau đó phục hồi nhưng lại suy giảm mạnh vào cuối chu kỳ. Điều này cho thấy lợi nhuận trên mỗi cổ phiếu của công ty không ổn định và có dấu hiệu suy yếu trong giai đoạn gần đây.
- Sự khác biệt trong xu hướng của BVPS và EPS có thể cho thấy công ty đang tích lũy tài sản nhưng hiệu quả sinh lời trên tài sản đó lại giảm sút.


```r
bid_cstc$ThoiGian <- as.Date(paste0(bid_cstc$Năm, "-", bid_cstc$Tháng, "-01"))

data_long <- bid_cstc %>%
  pivot_longer(cols = c(P.E),
              names_to = "Variable",
              values_to = "Value")

ggplot(data_long, aes(x = ThoiGian, y = Value, color = Variable)) +
  geom_line() +
  labs(title = "P/E theo thời gian") +
  theme_minimal()

data_long <- bid_cstc %>%
  pivot_longer(cols = c(P.B),
              names_to = "Variable",
              values_to = "Value")

ggplot(data_long, aes(x = ThoiGian, y = Value, color = Variable)) +
  geom_line() +
  labs(title = "P/B theo thời gian") +
  theme_minimal()

data_long <- bid_cstc %>%
  pivot_longer(cols = c(P.S),
              names_to = "Variable",
              values_to = "Value")

ggplot(data_long, aes(x = ThoiGian, y = Value, color = Variable)) +
  geom_line() +
  labs(title = "P/S theo thời gian") +
  theme_minimal()
```

P/E Ratio Over Time:
- Quý 1/2022 - Quý 2/2023: Tỷ lệ P/E giảm từ khoảng 16 xuống mức thấp nhất là khoảng 10 vào quý 2/2023.
- Quý 2/2023 - 2024: Tỷ lệ P/E có sự hồi phục nhẹ nhưng vào cuối năm 2023 tỷ lệ có sự giảm trở lại.

P/B Ratio Over Time:
- Quý 1/2022 - Quý 2/2022: Tỷ lệ P/B có sự biến động giảm.
- Quý 2/2022 - Quý 1/2023: Tỷ lệ P/B có xu hướng tăng về giá trị đầu năm, sau đó giá trị ổn định quanh mức 2.2.
- Quý 2/2023 - Quý 4/2023: Tỷ lệ P/B có có biến động mạnh, tỷ lệ giảm nhẹ vào giai đoạn đầu, sau đó tăng mạnh vào cuối kì.
- Năm 2024, Tỷ lệ P/B giảm mạnh và đạt đáy vào cuối năm 2024.

P/S Ratio Over Time:
- Tỷ lệ P/S có xu hướng tương đồng với P/B.
- Tỷ lệ P/S có sự biến động đáng kể trong giai đoạn khảo sát, với xu hướng chung là tăng trong giai đoạn đầu và giữa, sau đó có sự sụt giảm nhẹ vào cuối giai đoạn.

```r
bid_cstc$color <- ifelse(bid_cstc$P.Cash.Flow > 0, "green", "red")

ggplot(bid_cstc, aes(x = ThoiGian, y = P.Cash.Flow , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "P/Cash Flow theo thời gian",
    x = "Năm - Tháng",
    y = "P/Cash Flow"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

- Năm 2024: tỷ lệ P/Cash Flow tăng mạnh và đạt đỉnh vào đầu năm, nhưng sau đó có sự điều chỉnh xuống và ổn định.
