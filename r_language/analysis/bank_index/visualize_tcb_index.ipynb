## Kết quả kinh doanh

```r
tcb_kqkd$Tháng <- tcb_kqkd$Kỳ * 3
head(tcb_kqkd, 5)
```

```r
tcb_kqkd$ThoiGian <- as.Date(paste0(tcb_kqkd$Năm, "-", tcb_kqkd$Tháng, "-01"))

ggplot(tcb_kqkd, aes(x = ThoiGian, y = Doanh.thu..đồng.)) +
  geom_line(group = 1, color = "steelblue") +
  geom_point(color = "darkblue") +
  labs(
    title = "Doanh thu (đồng) theo thời gian",
    x = "Năm - Kỳ",
    y = "Doanh thu (đồng)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ, ta thấy:
- Từ đầu năm 2022 - 2025, doanh thu có xu hướng tăng và đạt đỉnh ở quý 2 năm 2024. Giai đoạn tăng mạnh nhất từ quý 2 2022 - quý 1 năm 2023.
- Sau quý 2 năm 2024, doanh thu có xu hướng giảm nhẹ.

```r
tcb_kqkd$color <- ifelse(tcb_kqkd$Tăng.trưởng.doanh.thu.... > 0, "green", "red")

ggplot(tcb_kqkd, aes(x = ThoiGian, y = Tăng.trưởng.doanh.thu.... , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Tăng trưởng doanh thu (%) theo thời gian",
    x = "Năm - Kỳ",
    y = "Tăng trưởng doanh thu (%)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ tăng trưởng doanh thu, ta thấy:
- Từ đầu năm 2022 - 2024, tăng trưởng doanh thu luôn dương. Trong đó, quý 2 năm 2023 tăng trưởng doanh thu đạt đỉnh.
- Trong năm 2022 - quý 2 2023, tăng trưởng doanh thu cao nhưng ổn định.
- Từ quý 2 năm 2023, tăng trưởng doanh thu có xu hướng giảm. Và đạt thấp nhất từ quý 4 năm 2024.

```r
tcb_kqkd$color <- ifelse(tcb_kqkd$Tăng.trưởng.lợi.nhuận.... > 0, "green", "red")

ggplot(tcb_kqkd, aes(x = ThoiGian, y = Tăng.trưởng.lợi.nhuận.... , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Tăng trưởng lợi nhuận (%) theo thời gian",
    x = "Năm - Kỳ",
    y = "Tăng trưởng lợi nhuận (%)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ tăng trưởng lợi nhuận, ta thấy:

## Chỉ số tài chính

```r
tcb_cstc$Tháng <- tcb_cstc$Kỳ * 3
head(tcb_cstc, 5)
```

```r
tcb_cstc$ThoiGian <- as.Date(paste0(tcb_cstc$Năm, "-", tcb_cstc$Tháng, "-01"))

scale_factor <- max(tcb_cstc$ROE...., na.rm = TRUE) / max(tcb_cstc$ROA...., na.rm = TRUE)

ggplot(tcb_cstc, aes(x = ThoiGian)) +
  geom_line(aes(y = ROE...., color = "ROE (%)"), size = 1) +
  geom_point(aes(y = ROE...., color = "ROE (%)")) +
  geom_line(aes(y = ROA.... * scale_factor, color = "ROA (%)"), size = 1) +
  geom_point(aes(y = ROA.... * scale_factor, color = "ROA (%)")) +
  scale_y_continuous(
    name = "ROE (%)",
    sec.axis = sec_axis(~ . / scale_factor, name = "ROA (%)")
  ) +
  scale_color_manual(
    name = "",
    values = c("ROE (%)" = "#1f77b4", "ROA (%)" = "#ff7f0e")
  ) +
  labs(
    title = "ROE và ROA theo thời gian",
    x = "Năm - Kỳ"
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    legend.position = "top"
  )
```

Đối với ROE:
- Quý 1 2022 - Quý 3 2023: ROE có xu hướng giảm đáng kể từ khoảng 21% xuống còn khoảng 14%.
- Quý 4 2023 - Quý 4 2024: ROE có dấu hiệu phục hồi và tăng trở lại.

Đối với ROA:
- Quý 1 2022 - Quý 4 2023: ROA có xu hướng giảm tương tự như ROE, từ khoảng 3.3% xuống còn 2.3%.
- Quý 1 2024 - Quý 4 2024: ROA phục hồi và tăng nhẹ. Nhưng và cuối năm, ROE quay đầu giảm trở lại.

So sánh ROE và ROA:
- Cả ROE và ROA đều có xu hướng giảm mạnh như nhau trong giai đoạn từ đầu năm 2022 đến đầu năm 2024.
- Đến giai đoạn cuối, cả hai chỉ số đều có dấu hiệu phục hồi, tuy nhiên mức độ phục hồi của ROE rõ rệt hơn so với ROA.
- ROE luôn cao hơn ROA, điều này là hợp lý vì ROE đo lường lợi nhuận trên vốn chủ sở hữu, trong khi ROA đo lường lợi nhuận trên tổng tài sản (bao gồm cả nợ). Sự khác biệt giữa ROE và ROA phản ánh tác động của đòn bẩy tài chính. Tuy vậy, ROE và ROA có tỉ lệ tương quan cao với nhau.

```r
tcb_cstc$color <- ifelse(tcb_cstc$Biên.lợi.nhuận.ròng.... > 0, "green", "red")

ggplot(tcb_cstc, aes(x = ThoiGian, y = Biên.lợi.nhuận.ròng.... , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Biên lợi nhuận ròng (%) theo thời gian",
    x = "Năm - Kỳ",
    y = "Biên lợi nhuận ròng (%)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Theo biểu đồ, ta thấy:
- Năm 2022: Biên lợi nhuận ở mức vừa, sau đó tăng cao nhất ở quý 2, rồi giảm mạnh ở cuối năm.
- Biên lợi nhuận có tính chu kỳ theo năm. Từ mức cao ở đầu năm và giảm dần về cuối năm.
- Quý 4/2024: Biên lợi nhuận ròng giảm xuống mức thấp nhất trong toàn bộ chu kỳ được thể hiện trên biểu đồ (khoảng 39%).

```r
tcb_cstc$ThoiGian <- as.Date(paste0(tcb_cstc$Năm, "-", tcb_cstc$Tháng, "-01"))

data_long <- tcb_cstc %>%
  pivot_longer(cols = c(Vốn.hóa..Tỷ.đồng.),
              names_to = "Variable",
              values_to = "Value")

ggplot(data_long, aes(x = ThoiGian, y = Value, color = Variable)) +
  geom_line() +
  labs(title = "Vốn hóa (tỷ đồng) theo thời gian", x = "Năm - Kỳ", y = "Vốn hóa (tỷ đồng)") +
  theme_minimal()
```

Theo biểu đồ, ta thấy:

```r
tcb_cstc$color <- ifelse(tcb_cstc$Số.CP.lưu.hành..Triệu.CP. > 0, "green", "red")

ggplot(tcb_cstc, aes(x = ThoiGian, y = Số.CP.lưu.hành..Triệu.CP. , fill = color)) +
  geom_col() +
  scale_fill_identity() +
  labs(
    title = "Số cổ phiếu lưu hành (triệu cổ phiếu) theo thời gian",
    x = "Năm - Kỳ",
    y = "Số cổ phiếu lưu hành (triệu cổ phiếu)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Từ biểu đồ, ta thấy số cổ phiếu lưu hành có 1 lần điều chỉnh vào quý 1 năm 2024.

```r
tcb_cstc$ThoiGian <- as.Date(paste0(tcb_cstc$Năm, "-", tcb_cstc$Tháng, "-01"))

scale_factor <- max(tcb_cstc$BVPS..VND., na.rm = TRUE) / max(tcb_cstc$EPS..VND., na.rm = TRUE)

ggplot(tcb_cstc, aes(x = ThoiGian)) +
  geom_col(aes(y = BVPS..VND., fill = "BVPS"), width = 25) +
  geom_line(aes(y = EPS..VND. * scale_factor, color = "EPS"), size = 1.2) +
  geom_point(aes(y = EPS..VND. * scale_factor, color = "EPS")) +
  scale_y_continuous(
    name = "BVPS (VND)",
    sec.axis = sec_axis(~ . / scale_factor, name = "EPS (VND)")
  ) +
  scale_fill_manual(values = c("BVPS" = "skyblue"), name = "") +
  scale_color_manual(values = c("EPS" = "darkorange"), name = "") +
  labs(
    title = "BVPS và EPS theo thời gian",
    x = "Thời gian"
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    legend.position = "top"
  )
```

Về BVPS:
- Quý 1/2022 - Quý 4/2023: BVPS có xu hướng tăng trưởng khá ổn định trong giai đoạn này. Từ khoảng 28 nghìn đồng vào đầu năm 2022, BVPS tăng dần qua các quý và đạt mức cao nhất khoảng 37 nghìn đồng vào cuối năm 2023.
- Quý 1/2024 - Quý 4/2025: BVPS giảm xuống mức 19 nghìn đồng, sau đó có dấu hiệu đi ngang và duy trì ở mức tương đương, dao động nhẹ quanh mốc 20 nghìn đồng.

Về EPS:
- Năm 2022: EPS có sự biến động lớn trong giai đoạn này. Ban đầu ở mức khá cao, EPS giảm mạnh xuống mức thấp nhất vào cuối năm 2022.
- Năm 2023: EPS cho thấy sự phục hồi rõ rệt và có xu hướng ổn định hơn, dao động quanh mức 1250-1300.
- Từ đầu năm 2024, EPS bắt đầu có xu hướng giảm dần và giảm mạnh vào cuối năm 2024.

So sánh BVPS và EPS:
- BVPS cho thấy sự tăng trưởng ổn định trong phần lớn giai đoạn, cho thấy giá trị tài sản ròng trên mỗi cổ phiếu của công ty đang tăng lên.
- EPS lại có sự biến động mạnh hơn. Giai đoạn đầu có sự sụt giảm đáng kể, sau đó phục hồi nhưng lại suy giảm mạnh vào cuối chu kỳ. Điều này cho thấy lợi nhuận trên mỗi cổ phiếu của công ty không ổn định và có dấu hiệu suy yếu trong giai đoạn gần đây.
- Sự khác biệt trong xu hướng của BVPS và EPS có thể cho thấy công ty đang tích lũy tài sản nhưng hiệu quả sinh lời trên tài sản đó lại giảm sút.
