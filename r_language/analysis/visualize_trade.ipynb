# Visualize

## Each bank

```r
visualize_trade_each_bank <- function(bank) {
    data_long <- bank %>%
      pivot_longer(cols = c(Open, High, Low, Close),
                 names_to = "Variable",
                 values_to = "Value")

    ggplot(data_long, aes(x = Time, y = Value, color = Variable)) +
      geom_line() +
      labs(title = "Biểu đồ giá trị theo thời gian", x = "Thời gian", y = "Giá trị") +
      theme_minimal()
}
```

```r
visualize_volume_each_bank <- function(bank) {
    data_long <- bank %>%
      pivot_longer(cols = c(Volume),
                 names_to = "Variable",
                 values_to = "Value")

    ggplot(data_long, aes(x = Time, y = Value, color = Variable)) +
      geom_line() +
      labs(title = "Khối lượng giao dịch theo thời gian", x = "Thời gian", y = "Khối lượng giao dịch") +
      theme_minimal()
}
```

### Vietcombank

```r
visualize_trade_each_bank(vcb_trade)
```

Theo biểu đồ nến, ta thấy:
  - Giá cổ phiếu khá ổn định, từ năm 2022 có xu hướng hơi giảm, thấp nhất là đầu tháng 10.
  - Từ cuối năm 2022, giá cổ phiếu có xu hướng tăng, sau đó ổn định và đi ngang đến 2025.

## Three banks

```r
vcb_trade_new <- vcb_trade
colnames(vcb_trade_new) <- paste0("VCB_", colnames(vcb_trade))
colnames(vcb_trade_new)[colnames(vcb_trade_new) == "VCB_Time"] <- "Time"
head(vcb_trade_new, 5)
```

```r
bid_trade_new <- bid_trade
colnames(bid_trade_new) <- paste0("BID_", colnames(bid_trade))
colnames(bid_trade_new)[colnames(bid_trade_new) == "BID_Time"] <- "Time"
head(bid_trade_new, 5)
```

```r
tcb_trade_new <- tcb_trade
colnames(tcb_trade_new) <- paste0("TCB_", colnames(tcb_trade))
colnames(tcb_trade_new)[colnames(tcb_trade_new) == "TCB_Time"] <- "Time"
head(tcb_trade_new, 5)
```

```r
visualize_close_three_bank <- function() {
    banks = merge(vcb_trade_new, bid_trade_new, by.x="Time")
    banks = merge(banks, tcb_trade_new, by.x="Time")

    data_long <- banks %>%
      pivot_longer(cols = ends_with("Close"),
                 names_to = "Variable",
                 values_to = "Value")

    ggplot(data_long, aes(x = Time, y = Value, color = Variable)) +
      geom_line() +
      labs(title = "Biểu đồ giá trị theo thời gian", x = "Thời gian", y = "Giá trị") +
      theme_minimal()
}
```

```r
visualize_close_three_bank()
```
