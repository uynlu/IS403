## CPI

```r
cpi$Time <- as.Date(cpi$Time, origin = "1899-12-30")
head(cpi, 5)
```

```r
cpi$Time <- cpi$Time %m+% months(1)
cpi$Time <- format(cpi$Time, "%Y-%m")
head(cpi, 5)
```

```r
names(cpi)[names(cpi) == "Value"] <- "Previous_CPI"
head(cpi, 5)
```

## GDP

```r
gdp <- gdp %>%
  mutate(
    Quarter = substr(Time, 1, 2),

    Year = as.numeric(substr(Time, 4, 7)),

    Month = case_when(
      Quarter == "Q1" ~ "01",
      Quarter == "Q2" ~ "04",
      Quarter == "Q3" ~ "07",
      Quarter == "Q4" ~ "10"
    ),

    Date = as.Date(paste0(Year, "-", Month, "-01"))
  )

head(gdp, 5)
```

```r
gdp$Date <- gdp$Date %m+% months(3)
head(gdp, 5)
```

```r
gdp <- gdp %>%
  mutate(
    Quarter = quarter(Date),
    Year = year(Date),
    QuarterLabel = paste0("Q", Quarter, " ", Year)
  )
head(gdp, 5)
```

```r
gdp <- gdp[, !names(gdp) %in% c('Time', 'Quarter', 'Year', 'Month', 'Date')]
names(gdp)[names(gdp) == "QuarterLabel"] <- "Time"
head(gdp, 5)
```

```r
names(gdp)[names(gdp) == "Value"] <- "Previous_GDP"
head(gdp, 5)
```

## Foreign exchange rate

```r
tcb_foreign_exchange_rate$Date <- as.Date(tcb_foreign_exchange_rate$Date, format = "%Y-%m-%d")
head(tcb_foreign_exchange_rate, 5)
```

```r
tcb_foreign_exchange_rate$Date <- tcb_foreign_exchange_rate$Date + days(1)
head(tcb_foreign_exchange_rate, 5)
```

```r
colnames(tcb_foreign_exchange_rate) <- sapply(colnames(tcb_foreign_exchange_rate), function(x) {
  if(x != "Date") {
    paste("Previous_", x)
  } else {
    x
  }
})

head(tcb_foreign_exchange_rate, 5)
```

## Bank index

### Kết quả kinh doanh

```r
tcb_kqkd <- tcb_kqkd %>%
  mutate(
    Ngày = ymd(paste0(
      Năm,
      "-",
      case_when(
        Kỳ == 1 ~ "01",
        Kỳ == 2 ~ "04",
        Kỳ == 3 ~ "07",
        Kỳ == 4 ~ "10"
      ),
      "-01"
    ))
  )
head(tcb_kqkd, 5)
```

```r
tcb_kqkd$Ngày <- tcb_kqkd$Ngày %m+% months(3)
tcb_kqkd$Ngày <- format(tcb_kqkd$Ngày, "%Y-%m")
head(tcb_kqkd, 5)
```

```r
tcb_kqkd <- tcb_kqkd %>%
  mutate(
    Ngày = ym(Ngày),
    Năm = year(Ngày),
    Kỳ = quarter(Ngày)
  )
head(tcb_kqkd, 5)
```

```r
tcb_kqkd <- tcb_kqkd[, !names(tcb_kqkd) %in% c("Ngày")]
head(tcb_kqkd, 5)
```

```r
tcb_kqkd$Quarter <- paste0("Q", tcb_kqkd$Kỳ, " ", tcb_kqkd$Năm)
head(tcb_kqkd, 5)
```

## Trade

```r
tcb_trade$Time <- as.Date(tcb_trade$Time, format = "%Y-%m-%d")
head(tcb_trade, 5)
```

```r
tcb_trade <- tcb_trade %>%
  mutate(
    quarter = paste0("Q", quarter(Time)),
    year = year(Time),
    quarter_year = paste0(quarter, " ", year)
  ) %>%
  select(-quarter, -year)
head(tcb_trade, 5)
```

```r
tcb_trade$Month <- format(tcb_trade$Time, "%Y-%m")
head(tcb_trade, 5)
```

# Merge data

```r
tcb_dataset <- merge(tcb_trade, gdp, by.x="quarter_year", by.y="Time", all=TRUE)
head(tcb_dataset, 5)
```

```r
tcb_dataset <- merge(tcb_dataset, cpi, by.x="Month", by.y="Time", all=TRUE)
tcb_dataset <- tcb_dataset[, !names(tcb_dataset) %in% c("Month")]
head(tcb_dataset, 5)
```

```r
tcb_dataset <- merge(tcb_dataset, tcb_foreign_exchange_rate, by.x="Time", by.y="Date", all.x=TRUE)
tcb_dataset <- tcb_dataset[, !names(tcb_dataset) %in% c("Date")]
head(tcb_dataset, 5)
```

```r
tcb_dataset <- merge(tcb_dataset, tcb_kqkd, by.x="quarter_year", by.y="Quarter", all.x=TRUE)
tcb_dataset <- tcb_dataset[, !names(tcb_dataset) %in% c("Năm", "Kỳ")]
head(tcb_dataset, 5)
```

```r
tcb_dataset <- merge(tcb_dataset, tcb_cstc, by.x="quarter_year", by.y="Quarter", all.x=TRUE)
tcb_dataset <- tcb_dataset[, !names(tcb_dataset) %in% c("Năm", "Kỳ", "CP")]
head(tcb_dataset, 5)
```

```r
tcb_dataset <- merge(tcb_dataset, tcb_cdtc, by.x="quarter_year", by.y="Quarter", all.x=TRUE)
tcb_dataset <- tcb_dataset[, !names(tcb_dataset) %in% c("Năm", "Kỳ", "Quarter", "quarter_year")]
head(tcb_dataset, 5)
```
