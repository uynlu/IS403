## CPI

```r
cpi$Time <- as.Date(cpi$Time, origin = "1899-12-30")
head(cpi, 5)
```

```r
cpi$Time <- cpi$Time %m+% months(1)
cpi$Time <- format(cpi$Time, "%Y-%m")
head(cpi, 5)
```

```r
names(cpi)[names(cpi) == "Value"] <- "Previous_CPI"
head(cpi, 5)
```

## GDP

```r
gdp <- gdp %>%
  mutate(
    Quarter = substr(Time, 1, 2),
    Year = as.numeric(substr(Time, 4, 7)),

    Month = case_when(
      Quarter == "Q1" ~ "01",
      Quarter == "Q2" ~ "04",
      Quarter == "Q3" ~ "07",
      Quarter == "Q4" ~ "10"
    ),

    Date = as.Date(paste0(Year, "-", Month, "-01"))
  )

head(gdp, 5)
```

```r
gdp$Date <- gdp$Date %m+% months(3)
head(gdp, 5)
```

```r
gdp <- gdp %>%
  mutate(
    Quarter = quarter(Date),
    Year = year(Date),
    QuarterLabel = paste0("Q", Quarter, " ", Year)
  )
head(gdp, 5)
```

```r
gdp <- gdp[, !names(gdp) %in% c('Time', 'Quarter', 'Year', 'Month', 'Date')]
names(gdp)[names(gdp) == "QuarterLabel"] <- "Time"
head(gdp, 5)
```

```r
names(gdp)[names(gdp) == "Value"] <- "Previous_GDP"
head(gdp, 5)
```

## Foreign exchange rate

```r
vcb_foreign_exchange_rate$Date <- as.Date(vcb_foreign_exchange_rate$Date, format = "%Y-%m-%d")
head(vcb_foreign_exchange_rate, 5)
```

```r
vcb_foreign_exchange_rate$Date <- vcb_foreign_exchange_rate$Date + days(1)
head(vcb_foreign_exchange_rate, 5)
```

```r
colnames(vcb_foreign_exchange_rate) <- sapply(colnames(vcb_foreign_exchange_rate), function(x) {
  if(x != "Date") {
    paste("Previous_", x)
  } else {
    x
  }
})

head(vcb_foreign_exchange_rate, 5)
```

## Bank index

### Kết quả kinh doanh

```r
vcb_kqkd <- vcb_kqkd %>%
  mutate(
    Ngày = ymd(paste0(
      Năm,
      "-",
      case_when(
        Kỳ == 1 ~ "01",
        Kỳ == 2 ~ "04",
        Kỳ == 3 ~ "07",
        Kỳ == 4 ~ "10"
      ),
      "-01"
    ))
  )
head(vcb_kqkd, 5)
```

```r
vcb_kqkd$Ngày <- vcb_kqkd$Ngày %m+% months(3)
vcb_kqkd$Ngày <- format(vcb_kqkd$Ngày, "%Y-%m")
head(vcb_kqkd, 5)
```

```r
vcb_kqkd <- vcb_kqkd %>%
  mutate(
    Ngày = ym(Ngày),
    Năm = year(Ngày),
    Kỳ = quarter(Ngày)
  )
head(vcb_kqkd, 5)
```

```r
vcb_kqkd <- vcb_kqkd[, !names(vcb_kqkd) %in% c("Ngày")]
head(vcb_kqkd, 5)
```

```r
vcb_kqkd$Quarter <- paste0("Q", vcb_kqkd$Kỳ, " ", vcb_kqkd$Năm)
head(vcb_kqkd, 5)
```

## Trade

```r
vcb_trade$Time <- as.Date(vcb_trade$Time, format = "%Y-%m-%d")
head(vcb_trade, 5)
```

```r
vcb_trade <- vcb_trade %>%
  mutate(
    quarter = paste0("Q", quarter(Time)),
    year = year(Time),
    quarter_year = paste0(quarter, " ", year)
  ) %>%
  select(-quarter, -year)
head(vcb_trade, 5)
```

```r
vcb_trade$Month <- format(vcb_trade$Time, "%Y-%m")
head(vcb_trade, 5)
```

# Merge data

```r
vcb_dataset <- merge(vcb_trade, gdp, by.x="quarter_year", by.y="Time", all=TRUE)
head(vcb_dataset, 5)
```

```r
vcb_dataset <- merge(vcb_dataset, cpi, by.x="Month", by.y="Time", all=TRUE)
vcb_dataset <- vcb_dataset[, !names(vcb_dataset) %in% c("Month")]
head(vcb_dataset, 5)
```

```r
vcb_dataset <- merge(vcb_dataset, vcb_foreign_exchange_rate, by.x="Time", by.y="Date", all.x=TRUE)
vcb_dataset <- vcb_dataset[, !names(vcb_dataset) %in% c("Date")]
head(vcb_dataset, 5)
```

```r
vcb_dataset <- merge(vcb_dataset, vcb_kqkd, by.x="quarter_year", by.y="Quarter", all.x=TRUE)
vcb_dataset <- vcb_dataset[, !names(vcb_dataset) %in% c("Năm", "Kỳ")]
head(vcb_dataset, 5)
```

```r
vcb_dataset <- merge(vcb_dataset, vcb_cstc, by.x="quarter_year", by.y="Quarter", all.x=TRUE)
vcb_dataset <- vcb_dataset[, !names(vcb_dataset) %in% c("Năm", "Kỳ", "CP")]
head(vcb_dataset, 5)
```

```r
vcb_dataset <- merge(vcb_dataset, vcb_cdtc, by.x="quarter_year", by.y="Quarter", all.x=TRUE)
vcb_dataset <- vcb_dataset[, !names(vcb_dataset) %in% c("Năm", "Kỳ", "Quarter", "quarter_year")]
head(vcb_dataset, 5)
```
