### Correlation

#### Between features

```r
visualize_heatmap_corr <- function(dataframe) {
    library(ggcorrplot)
    library(ggplot2)

    corr_matrix <- cor(dataframe, use = "complete.obs")
    var_order <- rownames(corr_matrix)

    ggcorrplot(corr_matrix,
               method = "square",
               type = "lower",
               lab = TRUE,
               lab_size = 3,
               color = c("blue", "white", "red")) +
    scale_x_discrete(limits = var_order)          +
    scale_y_discrete(limits = rev(var_order))     +
    ggtitle("Mối tương quan giữa các biến")       +
    theme(plot.title = element_text(hjust = 0.5))
}

find_pair_features_high_corr <- function(dataframe) {
    corr_matrix <- cor(dataframe, use = "complete.obs")
    corr_df <- as.data.frame(as.table(corr_matrix))
    colnames(corr_df) <- c("Feature1", "Feature2", "Correlation")

    corr_df <- corr_df[corr_df$Feature1 != corr_df$Feature2, ]
    corr_df <- corr_df[as.character(corr_df$Feature1) < as.character(corr_df$Feature2), ]

    high_corr_pairs <- subset(corr_df, abs(Correlation) > 0.9)

    return(high_corr_pairs)
}
```

```r
options(repr.plot.width = 20, repr.plot.height = 20)
visualize_heatmap_corr(vcb[, !names(vcb) %in% c("Time", "Close")])
```

```r
options(repr.plot.width = 20, repr.plot.height = 20)
visualize_heatmap_corr(bid[, !names(bid) %in% c("Time", "Close")])
```

```r
visualize_heatmap_corr_target <- function(dataframe) {

    corr_matrix <- cor(dataframe, use = "complete.obs")

    cor_close <- as.data.frame(corr_matrix[, "Close", drop = FALSE])
    cor_close$Variable <- rownames(cor_close)
    colnames(cor_close)[1] <- "Correlation"

    ggplot(cor_close, aes(x = "Close", y = reorder(Variable, Correlation), fill = Correlation)) +
        geom_tile(color = "white") +
        geom_text(aes(label = round(Correlation, 2)), size = 3) +
        scale_fill_gradient2(low = "blue", mid = "white", high = "red", midpoint = 0, limits = c(-1, 1)) +
        labs(title = "Tương quan giữa các biến và Close", x = "", y = "") +
        theme_minimal() +
        theme(plot.title = element_text(hjust = 0.5))
}
```

## Random Forest

```r
target <- vcb$Close

rf_model <- randomForest(target ~ ., data = vcb, importance = TRUE, ntree = 100)

importance(rf_model)

varImpPlot(rf_model)
```
