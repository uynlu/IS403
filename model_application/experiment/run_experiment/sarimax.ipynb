# Excute SARIMAX

```python
def run_sarimax(train, test, checkpoint_path):
    start_time = time.time()
    model = pm.auto_arima(
        y=train["Close"],
        X=train.drop(columns=["Close", "Time"]),
        seasonal=True,
        stepwise=True, 
        trace=True
    )
    end_time = time.time()

    training_time = end_time - start_time
    json.dump(
        {"training time": training_time},
        open(os.path.join(checkpoint_path, f"training_time.json"), "w+"),
        ensure_ascii=False,
        indent=4
    )

    joblib.dump(model, open(os.path.join(checkpoint_path, f"sarimax.pkl"), "wb"))
    
    start_time = time.time()
    predictions = model.predict(
        n_periods=len(test),
        X=test.drop(columns=["Close", "Time"])
    )
    end_time = time.time()

    mse, rmse, mae, mape = error(predictions, test["Close"])
    elapsed = end_time - start_time

    json.dump(
        {
            "time": elapsed,
            "predictions": predictions.tolist(),
            "targets": test["Close"].tolist(),
            "mse": mse,
            "rmse": rmse,
            "mae": mae,
            "mape": mape
        },
        open(os.path.join(checkpoint_path, f"test_results.json"), "w+"),
        ensure_ascii=False,
        indent=4
    )

    return model
```

## Vietcombank

```python
vcb_checkpoint_path = ".\\model_application\\experiment\\saved_checkpoint\\arimax\\vcb"
```

```python
vcb_sarimax = run_sarimax(vcb_train, vcb_test, vcb_checkpoint_path)
```
