# Process data

```python
def scale(scaler_path, future):
    features_scaler = joblib.load(os.path.join(scaler_path, "features_scaler.pkl"))
    scaler = joblib.load(os.path.join(scaler_path, "scaler.pkl"))

    features = future.drop(columns=["Time", "Close"]).columns
    new_features_dataframe = pd.DataFrame(
        features_scaler.transform(future.drop(columns=["Time", "Close"])),
        columns=features
    )
    new_close_dataframe = pd.DataFrame(
        scaler.transform(future[["Close"]]),
        columns=["Close"]
    )
    return pd.concat([new_features_dataframe, new_close_dataframe], axis=1)
```

```python
vcb_scaler_path = ".\\model_application\\experiment\\saved_checkpoint\\deep_learning\\scaler\\vcb"
bid_scaler_path = ".\\model_application\\experiment\\saved_checkpoint\\deep_learning\\scaler\\bid"
tcb_scaler_path = ".\\model_application\\experiment\\saved_checkpoint\\deep_learning\\scaler\\tcb"
```

```python
vcb_future = scale(vcb_scaler_path, vcb_future)
vcb_future
```

```python
bid_future = scale(bid_scaler_path, bid_future)
bid_future
```

```python
tcb_future = scale(tcb_scaler_path, tcb_future)
tcb_future
```

# Predict

```python
N_STEPS = 4
```

```python
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
LEARNING_RATE = 5e-4
INPUT_SIZE = 9
HIDDEN_SIZE = 50
NUM_LAYERS = 9
DROPOUT_PROB = 0.3
NUM_EPOCHS = 100
USE_AMP = False
BATCH_SIZE = 16
```

## LSTM

### Vietcombank

```python
vcb_lstm_checkpoint_path = ".\\model_application\\experiment\\saved_checkpoint\\deep_learning\\lstm\\vcb\\multi_steps"
```

```python
vcb_lstm_executor = ModelExecutor(
    device=DEVICE,
    learning_rate=LEARNING_RATE,
    checkpoint_path=vcb_lstm_checkpoint_path,
    scaler_path=vcb_scaler_path,
    model_type="LSTM",
    input_size=INPUT_SIZE,
    hidden_size=HIDDEN_SIZE,
    num_layers=NUM_LAYERS,
    dropout_prob=DROPOUT_PROB,
    n_steps=N_STEPS,
    num_epochs=NUM_EPOCHS,
    use_amp=USE_AMP,
    batch_size=BATCH_SIZE
)
```

```python
vcb_lstm_executor.get_predictions(torch.Tensor(vcb_future.values).unsqueeze(0))
```
