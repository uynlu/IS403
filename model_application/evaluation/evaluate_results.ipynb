# Visualize

```python
def visualize_results(result, model):
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=pd.DataFrame(result["targets"]).index,
        y=result["targets"],
        mode="lines",
        name="Targets"
    ))
    fig.add_trace(go.Scatter(
        x=pd.DataFrame(result["predictions"]).index,
        y=result["predictions"],
        mode="lines",
        name="Predictions"
    ))

    fig.update_layout(
        title=f"{model} Predictions",
        xaxis_title="Time",
        yaxis_title="Close",
        xaxis=dict(showticklabels=False)
    )

    fig.show()
```

# Evaluate results

```python
def evaluate_results(data, error):
    model = smf.ols(f"{error} ~ Model", data=data).fit()
    anova_result = anova_lm(model)
    print(anova_result)
    print("\n")
    if anova_result["PR(>F)"]["Model"] < 0.02:
        tukey = pairwise_tukeyhsd(data[error], data["Model"]).summary().data
        tukey_dataframe = pd.DataFrame(tukey[1:], columns=tukey[0])
        print(tukey_dataframe[tukey_dataframe["reject"] == True].reset_index())
    else:
        print("Không có sự khác biệt giữa các chỉ số!")
```
