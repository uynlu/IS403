## Unusable currencies

```python
vcb_currencies = np.unique(vcb_foreign_exchange_rate["CurrencyCode"])
bid_currencies = np.unique(bid_foreign_exchange_rate["CurrencyCode"])
tcb_currencies = np.unique(tcb_foreign_exchange_rate["CurrencyCode"])

pd.set_option("display.max_colwidth", None)
pd.DataFrame([{
    "Bank": "Vietcombank",
    "Foreign currency": ", ".join(vcb_currencies),
    "Number of foreign currency": len(vcb_currencies)
}, {
    "Bank": "BIDV",
    "Foreign currency": ", ".join(bid_currencies),
    "Number of foreign currency": len(bid_currencies)
}, {
    "Bank": "Techcombank",
    "Foreign currency": ", ".join(tcb_currencies),
    "Number of foreign currency": len(tcb_currencies)
}])
```

```python
common_currencies = list(set(vcb_currencies) & set(bid_currencies) & set(tcb_currencies))
print("Những ngoại tệ mà cả 3 ngân hàng đều có:", common_currencies)
```

Nhận xét:
  - Cả 3 ngân hàng đều có USD, lấy giá trị của 10 USD.
  - Từ những ngoại tệ mà cả 3 ngân hàng đều có => Loại bỏ những ngoại tệ còn lại.

```python
def drop_unusable_currencies(dataframe, bank):
    if bank == "Vietcombank":
        filtered_dataframe = dataframe[dataframe["CurrencyCode"].isin(common_currencies) | dataframe["CurrencyCode"].str.contains("USD")]
    else:
        usd_pattern = r"USD\s?\((?=.*10)(?!.*100)(\d+(?:[-,]\d+)*)\)"
        filtered_dataframe = dataframe[dataframe["CurrencyCode"].isin(common_currencies) | dataframe["CurrencyCode"].str.contains(usd_pattern)]
        filtered_dataframe = filtered_dataframe.replace({
            "USD(10-20)": "USD",
            "USD (5,10,20)": "USD",
        })
    return filtered_dataframe
```

```python
vcb_foreign_exchange_rate = drop_unusable_currencies(vcb_foreign_exchange_rate, "Vietcombank")
vcb_foreign_exchange_rate
```

```python
bid_foreign_exchange_rate = drop_unusable_currencies(bid_foreign_exchange_rate, "BIDV")
bid_foreign_exchange_rate
```

```python
tcb_foreign_exchange_rate = drop_unusable_currencies(tcb_foreign_exchange_rate, "Techcombank")
tcb_foreign_exchange_rate
```

## Structure of dataframe

```python
def change_structure(dataframe):
    dataframe_pivot = dataframe.melt(
        id_vars=["Date", "CurrencyCode"],
        value_vars=["Cash", "Transfer", "Sell"], 
        var_name="Type",
        value_name="Value"
    )
    dataframe_pivot["Currency_Type"] = dataframe_pivot["CurrencyCode"] + "_" + dataframe_pivot["Type"]
    
    dataframe_final = dataframe_pivot.pivot_table(
        index="Date",
        columns="Currency_Type",
        values="Value",
        aggfunc="first"
    )
    dataframe_final.reset_index(inplace=True)

    dataframe_final.columns.name = None
    dataframe_final = dataframe_final.loc[:, ~dataframe_final.columns.str.contains("Currency_Type")]

    return dataframe_final
```

```python
vcb_foreign_exchange_rate = change_structure(vcb_foreign_exchange_rate)
vcb_foreign_exchange_rate
```

```python
bid_foreign_exchange_rate = change_structure(bid_foreign_exchange_rate)
bid_foreign_exchange_rate
```

```python
tcb_foreign_exchange_rate = change_structure(tcb_foreign_exchange_rate)
tcb_foreign_exchange_rate
```
